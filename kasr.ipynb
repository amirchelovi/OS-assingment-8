```python
class kasr:

    def __init__(self, x=None, y=None):
        self.x = x
        self.y = y

    def sum(self, other):
        result = kasr()
        result.x = self.x*other.y + self.y*other.x
        result.y = self.y*other.y
        return result

    def sub(self, other):
        result = kasr()
        result.x = self.x*other.y - self.y*other.x
        result.y = self.y*other.y
        return result

    def multiple(self, other):
        result = kasr()
        result.x = self.x*other.x
        result.y = self.y*other.y
        return result


    def divide(self, other):
        result = kasr()
        result.x = self.x*other.y
        result.y = self.y*other.x
        return result



    def show(self):
        return str(self.x) + '/' + str(self.y)

firstKasr= list(map(int, input('enter kasr1: x/y\n').split('/'))) 
socondKasr= list(map(int, input('enter kasr2: x/y\n').split('/'))) 
a = kasr(firstKasr[0], firstKasr[1])
b = kasr(socondKasr[0], socondKasr[1])
if firstKasr[1]== 0 & socondKasr[1] == 0 :
    print("this value is not exist!")
else:
    print('sum:',(a.sum(b)).show(),'\nsub:',(a.sub(b)).show(),'\nmul:',(a.multiple(b)).show(),'\ndiv:',(a.divide(b)).show())
```
